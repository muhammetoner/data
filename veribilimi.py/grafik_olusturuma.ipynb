```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("HousingData_cleaned.csv")
df
```

```python
df['AGE_CAT'] = pd.cut(df['AGE'], bins=[0, 30, 70, 100], labels=["Yeni", "Orta", "Eski"])

# Her kategori iÃ§in ortalama fiyat
age_grouped = df.groupby('AGE_CAT')['MEDV'].mean()

plt.figure(figsize=(6, 5))
age_grouped.plot(kind='bar', color='orange')
plt.title("Ev YaÅŸÄ± Kategorisine GÃ¶re Ortalama Fiyat")
plt.xlabel("Ev YaÅŸÄ± Kategorisi(AGE)")
plt.ylabel("Ortalama Ev FiyatÄ±(MEDV)")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.show()
```

```python
rm_grouped = df.groupby(df['RM'].round(1))['MEDV'].mean()

plt.figure(figsize=(8, 5))
plt.plot(rm_grouped.index, rm_grouped.values, marker='o', linestyle='-', color='teal')
plt.title("Oda SayÄ±sÄ± ArttÄ±kÃ§a Ortalama Ev FiyatÄ±")
plt.xlabel("Oda SayÄ±sÄ± (YuvarlatÄ±lmÄ±ÅŸ)")
plt.ylabel("Ortalama Ev FiyatÄ±")
plt.grid(True)
plt.show()
```

```python
rm_grouped.to_csv("rm_vs_medv.csv", index=True)
```
