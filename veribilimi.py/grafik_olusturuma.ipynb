```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("HousingData_cleaned.csv")
df
```

```python
df['AGE_CAT'] = pd.cut(df['AGE'], bins=[0, 30, 70, 100], labels=["Yeni", "Orta", "Eski"])

# Her kategori için ortalama fiyat
age_grouped = df.groupby('AGE_CAT')['MEDV'].mean()

plt.figure(figsize=(6, 5))
age_grouped.plot(kind='bar', color='orange')
plt.title("Ev Yaşı Kategorisine Göre Ortalama Fiyat")
plt.xlabel("Ev Yaşı Kategorisi(AGE)")
plt.ylabel("Ortalama Ev Fiyatı(MEDV)")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.show()
```

```python
rm_grouped = df.groupby(df['RM'].round(1))['MEDV'].mean()

plt.figure(figsize=(8, 5))
plt.plot(rm_grouped.index, rm_grouped.values, marker='o', linestyle='-', color='teal')
plt.title("Oda Sayısı Arttıkça Ortalama Ev Fiyatı")
plt.xlabel("Oda Sayısı (Yuvarlatılmış)")
plt.ylabel("Ortalama Ev Fiyatı")
plt.grid(True)
plt.show()
```

```python
rm_grouped.to_csv("rm_vs_medv.csv", index=True)
```

```python
age_grouped.to_csv("age_vs_medv.csv", index=True)
```
